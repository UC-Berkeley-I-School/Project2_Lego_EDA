```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Individual Lego DataFrames
colors = pd.read_csv("data/colors.csv")
parts = pd.read_csv("data/parts.csv")
sets = pd.read_csv("data/sets.csv")
themes = pd.read_csv("data/themes.csv")
inventories = pd.read_csv("data/inventories.csv")
inventory_parts = pd.read_csv("data/inventory_parts.csv")
inventory_sets = pd.read_csv("data/inventory_sets.csv")
part_categories = pd.read_csv("data/part_categories.csv")
#Import IMDB data
imdb_titles = pd.read_csv("data/title.basics.tsv.gz", sep= "\t")
```

```python
df_list = [colors, parts, sets, themes, inventories, inventory_parts, inventory_sets, part_categories, imdb_titles]
var_names = ["colors", "parts", "sets", "themes", "inventories", "inventory_parts", "inventory_sets", "part_categories", "imdb_titles"]
source = ["lego", "lego", "lego", "lego", "lego", "lego", "lego", "lego", "imdb"]
rows = []
cols = []
for df in df_list:
    rows.append(df.shape[0])
    cols.append(df.shape[1])
data_shapes = pd.DataFrame({"DataFrames":var_names, "Source": source, "Row Count": rows, "Column Count":cols})
data_shapes
```
